## Intro to Index Funds
An index is created by dominant financial institutions as an asset class or financial instrument to track and reflect the performance of a particular market, asset class, sector, or investment strategy. 

Unlike individual securities, market indexes aren’t directly investable. Instead, most investors use index funds or ETFs, which are designed to track the performance of designated indexes, less fees.

## How Are They Constructed ? 
Most major indices are market-cap weighted — meaning bigger companies have more influence on the index's movement. So if Apple (massive market cap) drops 5%, it drags the S&P 500 down far more than if a smaller company drops 5%.
Some indices are price-weighted (like the Dow Jones), where literally the higher the stock price, the more influence it has — which is a somewhat arbitrary and outdated method.

## How Index Providers Make Money?
At the end of the day, an index is just a model representative or so-called "information". Index providers such as ICE, Bloomberg, Standard & Poor Global, etc... license their indices to financial product issuers (ETF companies, fund managers, derivatives exchanges). Every time someone buys a Vanguard S&P 500 ETF, Vanguard pays S&P Global a licensing fee. This is an enormous, high-margin business — MSCI for example has ~70% operating margins. 

The reasoning behind their high margins is because Index Issuers are not required to hold any shares, perform or manage transactions, etc... There is no settlement or clearing involved. All they do is provide a "model" or informational representative of the financial market to the public. 

## How Is An Index Traded ? 
An index can only be traded after it is licensed by Sell-Side Insitution. Where it has been transformed to either a form of ETF, Mutual Fund, Derivative Product such as futures of CFD. etc...

## Good To Know Stuff
1. Index providers create and maintain indexes 
- Each provider uses a unique, rules-based methodology to build their indexes. This methodology is used to define the scope of the index, such as which securities or financial instruments are included and their respective weightings.
2. The composition of indexes change frequently 
- During periodic reviews, index providers may make changes to their indexes. For example, a provider will remove securities that no longer meet the criteria for index inclusion as outlined in its construction methodology. 
3. Fund managers manage funds, not indexes. 
- At the end of the day, indexes are just financial tools that are separately maintained by their providers. 
4. An index is purely information. 
- An index is just a calculated number that tracks closely to it's undelyings. It is meant to be a "model" of understanding the complex world of finance. 
- The "tradable" layer only comes into existence when an institution licenses that information and wraps it into a financial instrument.
